```python
import pandas as pd
```

```python
# load datasets
df_testing = pd.read_csv('customer_churn_dataset-testing-master.csv')
df_training = pd.read_csv('customer_churn_dataset-training-master.csv')
```

```python
df_testing.head(5)
```

```python
df_training.head(5)
```

```python
list = [df_training, df_testing] #create list for a loop

for x in list:
    print(x.isna().sum()) # check for missing values
```

```python
for x in list:
    x.dropna(inplace= True) # drop empty rows
```

```python
for x in list:
    print(x.isna().sum()) # verify change
```

```python
#we dont need the customerid column
for x in list:
    x.drop('CustomerID', axis=1, inplace=True)
```

```python
df_training
```

```python
from sklearn.preprocessing import LabelEncoder # some values are not numerical so we need to make them numerical 

feat_encoder = LabelEncoder()
```

```python
cat_feats = df_training.select_dtypes(include=['object']).columns # Get categorical columns
print(cat_feats)
```

```python
# Loop through each categorical feature in the 'cat_feats' list
for cat in cat_feats:
    # Apply the LabelEncoder to transform the categorical column 'cat' into numeric labels.
    df_training[cat] = feat_encoder.fit_transform(df_training[cat].astype(str))
```

```python
df_training = df_training.astype(int) # we know df_training has decimals which we dont need
```

```python
df_training.head(5)
```

```python
# Loop through each categorical feature in the 'cat_feats' list
for cat in cat_feats:
    # Apply the LabelEncoder to transform the categorical column 'cat' into numeric labels.
    df_testing[cat] = feat_encoder.fit_transform(df_testing[cat].astype(str))
```

```python
df_testing.head(5)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scale_columns = ['Age', 'Tenure', 'Usage Frequency', 'Support Calls', 'Total Spend', 'Last Interaction']

df_testing[scale_columns] = scaler.fit_transform(df_testing[scale_columns])
df_training[scale_columns] = scaler.fit_transform(df_training[scale_columns])
```

```python
df=pd.concat([df_testing,df_training],axis=0) #merge datasets into 1 to make it easier
df.head()
```

```python
#save datasets so normalization is no longer needed for the other approaches
df.to_csv('df_normalized.csv', index=False)
```
